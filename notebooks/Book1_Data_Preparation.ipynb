```python
data['trade_dt_epoch_ms'] = pd.to_datetime(data['Timestamp'], unit='ms', errors='coerce')
data['trade_dt_epoch_s']  = pd.to_datetime(data['Timestamp'], unit='s',  errors='coerce')
```

```python
if 'Timestamp IST' in data.columns:
    data['trade_dt_ist'] = pd.to_datetime(data['Timestamp IST'], dayfirst=True, errors='coerce')
else:
    data['trade_dt_ist'] = pd.Series([pd.NaT]*len(data), index=data.index)
```

```python
fallback_cols = [c for c in ['created_at', 'timestamp_str', 'time'] if c in data.columns]
if fallback_cols:
    # try the first fallback column
    data['trade_dt_fallback'] = pd.to_datetime(data[fallback_cols[0]], errors='coerce')
else:
    data['trade_dt_fallback'] = pd.Series([pd.NaT]*len(data), index=data.index)
```

```python
data['trade_dt'] = data['trade_dt_ist'].combine_first(data['trade_dt_epoch_ms']) \
                                       .combine_first(data['trade_dt_epoch_s']) \
                                       .combine_first(data['trade_dt_fallback'])
```

```python
data['date_gran'] = data['trade_dt'].dt.floor('D')
data['date_key']  = data['date_gran'].dt.strftime('%Y-%m-%d')
```

```python
print("Data shape:", data.shape)

cols_to_check = ['Timestamp', 'Timestamp IST'] if 'Timestamp IST' in data.columns else ['Timestamp']
print("Data types for key columns:\n", data.dtypes[cols_to_check])

key_cols = ['Timestamp', 'Timestamp IST', 'trade_dt_epoch_ms', 'trade_dt_epoch_s',
            'trade_dt_ist', 'trade_dt', 'date_key']
key_cols = [c for c in key_cols if c in data.columns]

print("\nFirst 10 rows (key columns):")
display(data[key_cols].head(10))

missing_trade_dt = data['trade_dt'].isna().sum()
print(f"\nMissing 'trade_dt': {missing_trade_dt} out of {len(data)}")

if missing_trade_dt > 0:
    print("\nSample rows with missing 'trade_dt':")
    display(data[data['trade_dt'].isna()].head(10))
```

```python
data['trade_dt_epoch_ms'] = pd.to_datetime(data['Timestamp'], unit='ms', errors='coerce')
data['trade_dt_epoch_s']  = pd.to_datetime(data['Timestamp'], unit='s', errors='coerce')
data['trade_dt_ist'] = pd.to_datetime(data['Timestamp IST'], dayfirst=True, errors='coerce')
```

```python
print("Epoch-ms range:", data['trade_dt_epoch_ms'].min(), "to", data['trade_dt_epoch_ms'].max())
print("Epoch-s  range:", data['trade_dt_epoch_s'].min(), "to", data['trade_dt_epoch_s'].max())
print("IST-string range:", data['trade_dt_ist'].min(), "to", data['trade_dt_ist'].max())
```

```python
def plausible(series, start='2010-01-01', end='2035-12-31'):
    s, e = pd.to_datetime(start), pd.to_datetime(end)
    return series.between(s, e).mean()
```

```python
print("Plausible fraction ms:", plausible(data['trade_dt_epoch_ms']))
print("Plausible fraction s :", plausible(data['trade_dt_epoch_s']))
```

```python
mask_ist = data['trade_dt_ist'].notna()
data.loc[mask_ist, 'diff_days_ms_vs_ist'] = (data.loc[mask_ist, 'trade_dt_epoch_ms'] - data.loc[mask_ist, 'trade_dt_ist']).dt.total_seconds() / 86400
data.loc[mask_ist, 'diff_days_s_vs_ist']  = (data.loc[mask_ist, 'trade_dt_epoch_s'] - data.loc[mask_ist, 'trade_dt_ist']).dt.total_seconds() / 86400
```

```python
print("Median diff (epoch-ms â€“ IST):", data['diff_days_ms_vs_ist'].median(skipna=True))
print("Median diff (epoch-s â€“ IST): ", data['diff_days_s_vs_ist'].median(skipna=True))
```

```python
THRESH = 2
mask_large_diff = mask_ist & data['diff_days_ms_vs_ist'].abs().gt(THRESH)
print("Rows where epoch-ms disagrees with IST by >2 days:", mask_large_diff.sum())
```

```python
import pytz
```

```python
def plausible_fraction(series, start='2010-01-01', end='2035-12-31'):
    s, e = pd.to_datetime(start), pd.to_datetime(end)
    return series.between(s, e).mean()
```

```python
frac_ms = plausible_fraction(data['trade_dt_epoch_ms'])
frac_s  = plausible_fraction(data['trade_dt_epoch_s'])
print("Plausible fraction (ms):", frac_ms, " | (s):", frac_s)
```

```python
use_ms = frac_ms >= frac_s
data['trade_dt_epoch_best'] = data['trade_dt_epoch_ms'] if use_ms else data['trade_dt_epoch_s']
print("â†’ Using epoch as", "milliseconds" if use_ms else "seconds")
```

```python
k_tz = pytz.timezone('Asia/Kolkata')
data['trade_dt_ist_local'] = pd.to_datetime(data['Timestamp IST'], dayfirst=True, errors='coerce').dt.tz_localize(k_tz, ambiguous='NaT', nonexistent='NaT')
data['trade_dt'] = data['trade_dt_ist_local'].combine_first(pd.to_datetime(data['trade_dt_epoch_best'], utc=True, errors='coerce'))
```

```python
data['trade_dt_utc'] = data['trade_dt'].dt.tz_convert('UTC') #normalize
```

```python
data['date_key_utc'] = data['trade_dt_utc'].dt.strftime('%Y-%m-%d')
```

```python
data['date_key_ist'] = data['trade_dt_utc'].dt.tz_convert('Asia/Kolkata').dt.strftime('%Y-%m-%d')
```

```python
print("Trade datetime range (UTC):", data['trade_dt_utc'].min(), "â†’", data['trade_dt_utc'].max())
print("Unique daily keys (UTC):", data['date_key_utc'].nunique())
```

Merging datasets

```python
sentiment['date'] = pd.to_datetime(sentiment['date'], errors='coerce')
sentiment['date_key'] = sentiment['date'].dt.strftime('%Y-%m-%d')
```

```python
merged = data.merge(
    sentiment[['date_key', 'value', 'classification']],
    left_on='date_key_utc', right_on='date_key',
    how='left', validate='m:1'
)

print("Merged shape:", merged.shape)
display(merged[['Timestamp IST', 'trade_dt_utc', 'value', 'classification']].head())
```
